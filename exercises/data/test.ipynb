# ERT Prospection Strategy: Evaluating Sedimentation Depth vs. Topography

**Objective:**
This exercise guides you through designing an ERT survey to evaluate sedimentation depth in a valley, considering topography. You will use the `resipy` library to simulate and analyze the impact of your choices (number of electrodes, cable spacing, sequence, etc.) on the quality of your results.


---

## 1. Problem Statement

You are tasked with mapping the depth of sedimentary layers in a small valley. The valley has a V-shaped cross-section, with a stream at the bottom. The goal is to determine the thickness of the sediments (low resistivity) overlying the bedrock (high resistivity).

**Constraints:**
- The valley is 50m wide at the top and 10m deep.
- You have access to a 100m cable and a maximum of 48 electrodes.
- The terrain is uneven, with a slope of about 20°.


---

## 2. Key Choices in ERT Survey Design

### 2.1 Number of Electrodes
- **Trade-off:** More electrodes → higher resolution, but more time-consuming and complex setup.
- **Recommendation:** For a 50m profile, 24-36 electrodes (spacing 1.5-2m) are reasonable.

### 2.2 Cable Spacing
- **Trade-off:** Smaller spacing → better resolution, but less depth penetration.
- **Recommendation:** Use 2m spacing for a balance between resolution and depth.

### 2.3 Topography
- **Impact:** Ignoring topography can lead to artifacts in the inversion.
- **Recommendation:** Always include topography in your model.

### 2.4 Sequence
- **Trade-off:** Different sequences (Wenner, Schlumberger, Dipole-Dipole) have different sensitivities.
- **Recommendation:** Use a combination of Wenner-Schlumberger for good depth penetration and resolution.


---

## 3. Practical Exercise


### 3.1 Setup the Model

```python
import numpy as np
import matplotlib.pyplot as plt
import resipy

# Define topography (V-shaped valley)
x = np.linspace(0, 50, 100)
z = np.zeros_like(x)
z[25:75] = 10 - np.abs(x[25:75] - 37.5)  # Simple V-shape

# Define true resistivity model: sediment (10 Ohm.m) over bedrock (1000 Ohm.m)
res_model = np.ones((100, 100)) * 1000
res_model[z.astype(int):, :] = 10

# Plot the true model
plt.figure(figsize=(10, 5))
plt.imshow(res_model, extent=(0, 50, 10, 0), aspect='auto')
plt.colorbar(label='Resistivity (Ohm.m)')
plt.title("True Resistivity Model")
plt.xlabel("Distance (m)")
plt.ylabel("Depth (m)")
plt.show()
```

### 3.2 Design the Survey

```python
# Define electrode positions (2m spacing, 25 electrodes)
electrodes = np.linspace(0, 50, 25)
# Define topography at electrode positions
topo = np.interp(electrodes, x, z)

# Create a resipy survey
survey = resipy.Survey(electrodes, topo=topo)
# Add a Wenner-Schlumberger sequence
survey.add_wenner_schlumberger(skip=1)
```

### 3.3 Simulate Data and Invert

```python
# Simulate apparent resistivity
data = survey.simulate(res_model, shape=(100, 100), dx=0.5)
# Add 5% noise
data.add_noise(0.05)
# Invert the data
inv = resipy.Inversion(survey, data)
inv_result = inv.run()
```

### 3.4 Visualize Results

```python
# Plot the inverted model
plt.figure(figsize=(10, 5))
inv_result.plot_model()
plt.title("Inverted Resistivity Model")
plt.show()
```
